# 無から始めるkeras 4(重みのあれこれ)

```python
#ライブラリ
import numpy as np
from keras.utils import np_utils
from keras.models import Model
from keras.layers import Dense, Input

#GPUの仕様に関する設定
import tensorflow as tf
from keras import backend as K
config = tf.ConfigProto(
    gpu_options = tf.GPUOptions(
        visible_device_list = "1",
        allow_growth = True,
        per_process_gpu_memory_fraction = 0.3))
sess = tf.Session(config=config)
K.set_session(sess)
```

## データセットの生成

```python
#データセット
data = np.random.rand(1000,2)
labels = (np.sum(data, axis=1) > 1.0) * 2 -1 #ラベル(-1,1)
```

## ネットワークの構築

```python
#入力層
Inputs = Input(shape=(2,))
#隠れ層
hidden = Dense(20, activation = 'tanh')
#出力層
predictions = Dense(1, activation = 'tanh')
```

## 学習

```python
#compile: 最適化関数，損失関数，評価指標
model = Model(input=Inputs, output = predictions(hidden(Inputs)))
model.compile('adam', 'hinge', metrics=['accuracy'])
#学習
model.fit(data, labels, nb_epoch=300, validation_split=0.2)
```

## 未知データの予測

```python
test_data = np.random.rand(1000,2)
test_label = (np.sum(test_data, axis=1) > 0.5) * 2 - 1
predict = np.sign(model.predict(test_data).flatten())
sum(predict == test_label) /1000
```

## 重みの取得

```python
print(hidden.get_weights(),predictions.get_weights())
```

```python
model.predict(np.array([[0.50,0.51]]))
```

## 重みの設定

```python
predictions.set_weights([np.random.rand(20,1), np.random.rand(1,)])
test_data = np.random.rand(1000,2)
test_label = (np.sum(test_data, axis=1) > 0.5) * 2 - 1
predict = np.sign(model.predict(test_data).flatten())
sum(predict == test_label) /1000
```
